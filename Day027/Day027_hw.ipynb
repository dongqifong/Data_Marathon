### 丟一個銅板，丟了100次，出現正面 50 次的機率有多大。
(提示: 先想是哪一種分配，然後透過 python 語法進行計算)


```python
'''
你的答案 
'''
print('丟銅板的結果只有兩種，即正面或反面')
print('重複操作的bernoulli 實驗，其機率分佈為二項分配(binomial distribution))')
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
n=100
p=0.5
r=np.arange(0,n+1)
print('Binomial distribution when n=100, r.v=50 and p=0.5')
probs_50=stats.binom.pmf(50,n,p)
print('the probability is:',probs_50)

plt.scatter(x=[50],y=[probs_50],c=[1])
plt.text(x=50+2,y=probs_50,s='(r,p)=(50,{0})'.format(round(probs_50,3)))

probs=stats.binom.pmf(r,n,p)
plt.bar(r,probs)
plt.title('Binomial Distribution for r.v=0~100, n=100 and p=0.5')
plt.xlabel('Random variable')
plt.ylabel('Probability')
plt.show()
```
